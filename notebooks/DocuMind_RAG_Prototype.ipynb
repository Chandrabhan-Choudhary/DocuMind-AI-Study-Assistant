1. Textbook Processor

````python
def process_textbook():
    """Kaggle-compatible PDF loading"""
    try:
        pdf_path = "/kaggle/input/solar-data/solarSystemNotes.pdf"  # â† Update this path
        
        if not os.path.exists(pdf_path):
            raise FileNotFoundError(f"PDF not found at {pdf_path} - check Kaggle dataset setup")
            
        loader = PyPDFLoader(pdf_path)
        pages = loader.load_and_split()
        
        texts = [p.page_content for p in pages if p.page_content.strip()]
        
        # Initialize ChromaDB (same as before)
        client = chromadb.Client()
        client = chromadb.PersistentClient(path="/kaggle/working/chroma_db")
        embedding_func = embedding_functions.SentenceTransformerEmbeddingFunction(
            model_name="all-MiniLM-L6-v2"
        )
        
        
        collection = client.create_collection(
            name="textbook",
            embedding_function=embedding_func
        )
        
        collection.add(
            documents=texts,
            ids=[str(i) for i in range(len(texts))]
        )
        
        return collection
    
    except Exception as e:
        print(f"Error: {str(e)}")
        return None

textbook_db = process_textbook()
````

2. Core AI Functions

````python
class StudyAssistant:
    def __init__(self):
        self.model = genai.GenerativeModel('gemini-1.5-pro')
    
    def ask_question(self, question):
        """RAG-powered Q&A with basic error handling"""
        try:
            results = textbook_db.query(
                query_texts=[question],
                n_results=3
            )
            context = "\n".join(results['documents'][0])
            
            response = self.model.generate_content(
                f"Use this textbook context to answer clearly:\n{context}\nQuestion: {question}"
            )
            return response.text
        
        except Exception as e:
            return f"Failed to answer: {str(e)}"
    
    
    def generate_quiz(self, topic):
        """Improved quiz generation with JSON validation"""
        try:
            prompt = f"""Generate a 3-question quiz about {topic} in perfect JSON format (no markdown):
            {{
                "topic": "string",
                "questions": [
                    {{
                        "question": "string",
                        "options": ["a", "b", "c"],
                        "answer": "string",
                        "explanation": "string"
                    }}
                ]
            }}
            Important: 
            - Only return the raw JSON, no additional text
            - Ensure all quotes are double quotes
            - Include 3 questions with 3 options each"""
            
            response = self.model.generate_content(prompt)
            
            # Clean the response
            json_str = response.text
            json_str = re.sub(r'^```json\s*|\s*```$', '', json_str)  # Remove markdown
            json_str = json_str.strip()  # Remove whitespace
            
            quiz_data = json.loads(json_str)
            
            # Validate structure
            if not isinstance(quiz_data.get("questions"), list) or len(quiz_data["questions"]) != 3:
                raise ValueError("Invalid quiz format")
                
            return quiz_data
            
        except json.JSONDecodeError as e:
            return {"error": "Invalid JSON format", "response": response.text}
        except Exception as e:
            return {"error": str(e), "response": response.text if 'response' in locals() else None}
````

# Initialize assistant

````python
assistant = StudyAssistant()
````
